## 2.1 Approximation sets

__Definition 1__ : Approximation sets
>$A \subseteq Z$ be a set of objective vetors.  
A is called an approximation set   
if any element of A does not dominate or is not equal to any other objective vector in A  
>> that is, set of incomparable solutions
>
> The set of all approximation sets is denoted as $\Omega$


hard to define what is a good approximation  
but can compare between approximations

### 4 types of dominance
> $\succ\succ$ strictly dominates : z1 is better than z2 in all objectives  
> $\succ$ dominates : z1 is not worse than z2 in all objectives and better in at least one objective(one does not doimante itself)  
> $\rhd$ better : A $\rhd$ B. A weakly dominates B and A $\neq$ B  
> $\succeq$ weekly dominates : z1 is not worse than z2 in all objectives  
>|| incomparable : neither z1 weakly dominates z2 nor z2 weakly dominates z1

### __Definition 2__ : Quality indicator(measure)
>An m-ary quality indicator I is a function 
$I : \Omega^m \rightarrow \mathbb{R}$, which assigns each vector $(A1,A2, ..., Am)$ of m approximation sets a real value $I(A1,A2, ..., Am)$

not single number but rather a combination of different quality indicator  
ex. Van Veldhuizen: $I = (I_{GD}, I_S, I_{ONVG})$  
$I_{GD}$: avg euclidean distnace  
$I_S$: variance of distance  
$I_{ONVG}$ : number of elements in A

### __Definition 3__ : Comparison Method
> Let $A,B \in \Omega$ be two approximation sets, $I = (I_1, I_2, ..., I_k)$ a combination of qualitiy indicators, E: $\mathbb{R}^k \times \mathbb{R}^k \rightarrow \{false, true\}$ a boolean function which takes 2 real vectors of length k as arguments.

> If all indicators in I are unary, the comparison method C_{I,E} defined by I and E is a boolean function of the form
>
> $$ C_{I,E}(A,B) = E(I(A),I(B)) $$
>
> where $I(A') = (I_1(A'), I_2(A'), ..., I_k(A'))$ for $A' \in \Omega$

> If all indicators in I are binary, the comparison method C_{I,E} defined by I and E is a boolean function of the form
>
> $$ C_{I,E}(A,B) = E(I(A,B),I(B,A)) $$
>
> where $I(A', B') = (I_1(A', B'), I_2(A', B'), ..., I_k(A', B'))$ for $A', B' \in \Omega$

## 2.3 Linking Comparison Methods and Dominace Relations

>If $C_{I,E}(A,B)$ is a sufficient condition for $A \rhd B$,  
this comparison method can indicate that A is better than B.  
(= $C_{I,E}(A,B) \Rightarrow A \rhd B$)

>If $C_{I,E}(A,B)$ is a necessary condition for $A \rhd B$,  
this comparison method can indicate whether A is better than B.  
(= $A \rhd B \Rightarrow C_{I,E}(A,B)$)  


### __Definition 4__: Compatibility and completeness
> Let $\blacktriangleright$ be an arbitrary binary relation on approximation sets. 

> The comparison method $C_{I,E}(A,B)$ is denoted as $\blacktriangleright$-compatible  
> if either for any A,B $\in \Omega$,  
$C_{I,E}(A,B) \Rightarrow A \blacktriangleright B$ or $C_{I,E}(A,B) \Rightarrow A \blacktriangleright B$ 

> The comparison method $C_{I,E}(A,B)$ is denoted as $\blacktriangleright$-complete
> if either for any A,B $\in \Omega$,  
$A \blacktriangleright B \Rightarrow C_{I,E}(A,B)$ or B $\blacktriangleright A \Rightarrow C_{I,E}(A,B)$ 

### __Definition 5__ : Binary epsilon-indicator

> Suppose without loss of generality, a minimization problem wit n positive objectives, i.e. $Z \subseteq \mathbb{R}^{+^{n}}$.
>
>An objective vector $z^1 = (z^1_1, z^1_2, ..., z^1_n) \in Z$ is said to __$\epsilon$-dominate__ another objective function ,$z^2 = (z^2_1, z^2_2, ..., z^2_n) \in Z$, denoted by $z^1 \succeq z^2$ iff  
>
>$$\forall 1 \leq i \leq n : z^1_i \leq \epsilon \cdot z^2_i$$
>
> for a given $\epsilon > 0$, define binary  $\epsilon$-indicator $I_{\epsilon}$ as
>
>
>$I_{\epsilon}$(A,B) = $\underset{\epsilon \in \mathbb{R}}{inf\{\forall z^2 \in B \exists z^1 \in A : z^1 \succeq_{\epsilon} z^2\}}$
>
>for any two approximation sets $A,B \in \Omega$  

> $I_{\epsilon}$(A,B) = $\underset{z^2 \in B}{max} \; \underset{z^1 \in A}{min} \; \underset{1 \leq i \leq n} {max} \; \frac{z^1_i} {z^2_i}$

## Discussion
## 5.1. Summary of Result
* unary quality indicator in general are not capalbe of indicating whether an approximation set is better than another.
* Existing unary indicators at best infer that approximation set is not worse than another. Most may mislead.
* Binary indicators has less limitations.
* binary $\epsilon$ indicator can detect whether an approximation set is better than another. But only some binary indicators do the same.
* Binary indicator has bigger complexity. Quadratic increase.

## 5.2. Conclusion
* we need at least binary quality indicator to detect the quality of algorithm
* unary indicator can be used as filtering uncomparable sets. -> reduce the space of interest.
* Indicator is just preference information thus need to be expained in detail.
* $\epsilon$-indicator is quite cool since it has many desirable aspects : cheap to commpute, provides how much better.
* considering the stochasticity of multiobjective EA, statistical test should be considered in comparison. It can be included in the indicator function/Boolean function. Or make attainment function by Grunert Da onesca
