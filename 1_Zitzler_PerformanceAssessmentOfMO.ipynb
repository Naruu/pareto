## 2.1 Approximation sets

>$A \subseteq Z$ be a set of objective vetors A is called an approximation set if any element of A does not dominate or is not equal to any other objective vector in A. The set of all approximation sets is denoted as $\Omega$

> that is, set of incomparable solutions

### 4 types of dominacne
> strictly dominates : z1 is better than z2 in all objectives

> better : A $\rhd$ B. A weakly dominates B and A $\neq$ B  
> weekly dominates : z1 is not worse than z2 in all objectives

### __Definition 2__ : Quality indicator(measure)
>An m-ary quality indicator I is a function 
I : $\Omega \rightarrow \mathbb{R}$, which assigns each vector (A1,A2, ..., Am) of m approximation sets a real value I(A1,A2, ..., Am)

not single number rather a combinatioin of idfferent quality indicator

$I_{ONVG})$ : number of elements in A

### __Definition 3__ : Comparison Method
> Let $A,B \in \Omega$ be two approximation sets, $I = (I_1, I_2, ..., I_k)$ a combintion of qualitiy indicators, E: $\mathbb{R}^k \times \mathbb{R}^k \rightarrow \{false, true\}$ a boolean function which takes 2 real vectors of length k as arguments.

> If all indicators in I are unary, the comparison method C_{I,E} defined by I and E is a boolean function of the form
>
> $$ C_{I,E}(A,B) = E(I(A),I(B)) $$
>
> where $I(A') = (I_1(A'), I_2(A'), ..., I_k(A'))$ for $A' \in \Omega$

> If all indicators in I are binary, the comparison method C_{I,E} defined by I and E is a boolean function of the form
>
> $$ C_{I,E}(A,B) = E(I(A,B),I(B,A)) $$
>
> where $I(A', B') = (I_1(A', B'), I_2(A', B'), ..., I_k(A', B'))$ for $A', B' \in \Omega$

## 2.3 Linking Comparison Methods and Dominacne Relations

>If $C_{I,E}(A,B)$ is a sufficient condition for $A \rhd B$,  
this comparison method can indicate that A is better than B.  
(= $C_{I,E}(A,B) \Rightarrow A \rhd B$)

>If $C_{I,E}(A,B)$ is a necessary condition for $A \rhd B$,  
this comparison method can indicate whether A is better than B.  
(= $A \rhd B \Rightarrow C_{I,E}(A,B)$)  
>> Think in contrapositive.

### __Definition 4__: Compatibility and completeness
> Let $\blacktriangleright$ be an arbitrary binary relation on approximation sets. 

> The comparison method $C_{I,E}(A,B)$ is denoted as $\blacktriangleright$-compatible  
> if either for any A,B $\in \Omega$,  
$C_{I,E}(A,B) \Rightarrow A \blacktriangleright B$ or $C_{I,E}(A,B) \Rightarrow A \blacktriangleright B$ 

> The comparison method $C_{I,E}(A,B)$ is denoted as $\blacktriangleright$-complete
> if either for any A,B $\in \Omega$,  
$A \blacktriangleright B \Rightarrow C_{I,E}(A,B)$ or $\blacktriangleright B \Rightarrow A C_{I,E}(A,B)$ 

### __Definition 5: Binary $\epsilon$-indicator
> Suppose without loss of generality, a minimization problem wit n positive objectives, i.e. $Z \subseteq \mathbb{R}^{+^{n}}$.
>
>An objective vector $z^1 = (z^1_1, z^1_2, ..., z^1_n) \in Z$ is said to __$\epsilon$-dominate__ another objective function ,$z^2 = (z^2_1, z^2_2, ..., z^2_n) \in Z$, denoted by $z^1 \succeq z^2$ iff  
>
>$$\forall 1 \leq i \leq n : z^1_i \leq \epsilon \cdot z^2_i$$
>
> for a given $\epsilon > 0$, define binary  $\epsilon$-indicator $I_{\epsilon}$ as
>
>
>$I_{\epsilon}$(A,B) = $\underset{\epsilon \in \mathbb{R}}{inf\{\forall z^2 \in B \exists z^1 \in A : z^1 \succeq_{\epsilon} z^2\}}$
>
>for any two approximation sets $A,B \in \Omega$  

> $I_{\epsilon}$(A,B) = $\underset{z^2 \in B}{max} \; \underset{z^1 \in A}{min} \; \underset{1 \leq i \leq n} {max} \; \frac{z^1_i} {z^2_i}$

> $C_{I,E}(A,B) \Leftrightarrow A \rhd B$ for any approximation sets A,B $\in \Omega$$
